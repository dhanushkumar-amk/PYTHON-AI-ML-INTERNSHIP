### **Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Libraries imported!")
```

### **Load Dataset**

```python
df = pd.read_csv('train.csv')
print("Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

### **Basic Info**

```python
print("Total Rows    :", df.shape[0])
print("Total Columns :", df.shape[1])
print("\nData Types Count:\n", df.dtypes.value_counts())
print("\nColumn Names:\n", df.columns.tolist())
```

### **Statistical Summary**

```python
print("Numerical Summary:")
df.describe()
```

### **Missing Values**

```python
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
missing_pct = (missing / len(df) * 100).round(2)

missing_df = pd.DataFrame({'Missing Count': missing, 'Percentage %': missing_pct})
print(missing_df)
```

### **Visualize Missing Values**

```python
top_missing = missing.head(10)

plt.figure(figsize=(10, 4))
bars = plt.bar(top_missing.index, top_missing.values,
               color='tomato', edgecolor='black')
plt.title('Top 10 Columns with Missing Values')
plt.xlabel('Column')
plt.ylabel('Missing Count')
plt.xticks(rotation=45, ha='right')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 10,
             str(int(bar.get_height())),
             ha='center', fontsize=9)
plt.tight_layout()
plt.show()
```

### **Handle Missing Values**

```python
# Drop columns with more than 40% missing
threshold = 0.4 * len(df)
df.drop(columns=['PoolQC','MiscFeature','Alley','Fence','MasVnrType',
                 'FireplaceQu'], inplace=True)

# Fill numerical with median
df['LotFrontage'].fillna(df['LotFrontage'].median(), inplace=True)
df['MasVnrArea'].fillna(df['MasVnrArea'].median(), inplace=True)
df['GarageYrBlt'].fillna(df['GarageYrBlt'].median(), inplace=True)

# Fill categorical with mode
for col in ['GarageType','GarageFinish','GarageQual','GarageCond',
            'BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1',
            'BsmtFinType2','Electrical']:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("Missing after cleaning:\n", df.isnull().sum().sum())
print("Shape after cleaning:", df.shape)
```

### **Sale Price Distribution**

```python
plt.figure(figsize=(8, 4))
plt.hist(df['SalePrice'], bins=40, color='steelblue',
         edgecolor='black', alpha=0.8)
plt.title('Sale Price Distribution')
plt.xlabel('Sale Price ($)')
plt.ylabel('Number of Houses')
plt.axvline(df['SalePrice'].mean(), color='red', linestyle='--',
            label=f'Mean: ${df["SalePrice"].mean():,.0f}')
plt.axvline(df['SalePrice'].median(), color='orange', linestyle='--',
            label=f'Median: ${df["SalePrice"].median():,.0f}')
plt.legend()
plt.tight_layout()
plt.show()
```

### **Overall Quality vs Sale Price**

```python
qual_price = df.groupby('OverallQual')['SalePrice'].mean()

plt.figure(figsize=(8, 4))
bars = plt.bar(qual_price.index, qual_price.values,
               color='mediumpurple', edgecolor='black')
plt.title('Average Sale Price by Overall Quality')
plt.xlabel('Overall Quality (1=Poor, 10=Excellent)')
plt.ylabel('Average Sale Price ($)')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 2000,
             f'${int(bar.get_height()):,}',
             ha='center', fontsize=7, rotation=45)
plt.tight_layout()
plt.show()
```

### **Living Area vs Sale Price**

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['GrLivArea'], df['SalePrice'],
            color='darkorange', edgecolors='black', alpha=0.4, s=20)
plt.title('Living Area vs Sale Price')
plt.xlabel('Above Ground Living Area (sq ft)')
plt.ylabel('Sale Price ($)')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

### **Sale Price by Neighborhood**

```python
neigh_price = df.groupby('Neighborhood')['SalePrice'].median().sort_values(ascending=False)

plt.figure(figsize=(12, 5))
bars = plt.bar(neigh_price.index, neigh_price.values,
               color='teal', edgecolor='black')
plt.title('Median Sale Price by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Median Sale Price ($)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

### **Year Built vs Sale Price**

```python
year_price = df.groupby('YearBuilt')['SalePrice'].mean()

plt.figure(figsize=(10, 4))
plt.plot(year_price.index, year_price.values,
         color='blue', linewidth=1.5)
plt.title('Average Sale Price by Year Built')
plt.xlabel('Year Built')
plt.ylabel('Average Sale Price ($)')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

### **Correlation Heatmap**

```python
num_cols = ['SalePrice','OverallQual','GrLivArea','GarageCars',
            'TotalBsmtSF','1stFlrSF','FullBath','YearBuilt',
            'YearRemodAdd','LotArea']

corr = df[num_cols].corr().round(2)

fig, ax = plt.subplots(figsize=(9, 7))
im = ax.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(im)
ax.set_xticks(range(len(num_cols)))
ax.set_yticks(range(len(num_cols)))
ax.set_xticklabels(num_cols, rotation=45, ha='right', fontsize=8)
ax.set_yticklabels(num_cols, fontsize=8)
for i in range(len(corr)):
    for j in range(len(corr.columns)):
        ax.text(j, i, corr.iloc[i, j],
                ha='center', va='center', fontsize=7)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

### **Sale Price by Month Sold**

```python
month_price = df.groupby('MoSold')['SalePrice'].mean()
month_names = ['Jan','Feb','Mar','Apr','May','Jun',
               'Jul','Aug','Sep','Oct','Nov','Dec']

plt.figure(figsize=(9, 4))
plt.plot(month_names, month_price.values,
         marker='o', color='green', linewidth=2)
plt.title('Average Sale Price by Month Sold')
plt.xlabel('Month')
plt.ylabel('Average Sale Price ($)')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
print("=" * 50)
print("     HOUSE PRICES EDA - KEY FINDINGS")
print("=" * 50)
print(f"Total Houses        : {len(df)}")
print(f"Total Features      : {df.shape[1]}")
print(f"\nSale Price Mean     : ${df['SalePrice'].mean():,.0f}")
print(f"Sale Price Median   : ${df['SalePrice'].median():,.0f}")
print(f"Sale Price Min      : ${df['SalePrice'].min():,.0f}")
print(f"Sale Price Max      : ${df['SalePrice'].max():,.0f}")
print(f"\nTop Correlated with Price:")
corr_with_price = df[num_cols].corr()['SalePrice'].sort_values(ascending=False)
for col, val in corr_with_price.items():
    if col != 'SalePrice':
        print(f"  {col:20s}: {val:.2f}")
print("=" * 50)
```
