### **Import**

```python
import pandas as pd
print("Pandas version:", pd.__version__)
```

### **Series**

```python
s = pd.Series([100, 200, 300, 400, 500],
               index=['a', 'b', 'c', 'd', 'e'])
print(s)
print("\nAccess 'c':", s['c'])
print("Values > 200:\n", s[s > 200])
```

### **Creating DataFrame**

```python
data = {
    'Name'      : ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'Age'       : [24, 27, 22, 32, 29, None],
    'City'      : ['Chennai', 'Delhi', 'Mumbai', 'Kolkata', 'Pune', 'Delhi'],
    'Department': ['HR', 'IT', 'IT', 'Finance', 'HR', 'Finance'],
    'Salary'    : [45000, 52000, 38000, 61000, 55000, 48000],
    'Score'     : [88.5, 72.0, 91.0, 65.5, 80.0, None]
}
df = pd.DataFrame(data)
print(df)
```

### **Exploring DataFrame**

```python
print("Shape:", df.shape)
print("\nHead:\n", df.head(3))
print("\nTail:\n", df.tail(2))
print("\nInfo:")
df.info()
print("\nDescribe:\n", df.describe())
print("\nColumn Names:", df.columns.tolist())
```

### **Selecting Data**

```python
# Single column
print(df['Name'])

# Multiple columns
print(df[['Name', 'Salary', 'Department']])

# By row index (iloc)
print("Row 2 (iloc):\n", df.iloc[2])

# By condition (loc)
print("\nSalary > 50000:\n", df.loc[df['Salary'] > 50000])
```

### **Handling Missing Values**

```python
print("Null counts:\n", df.isnull().sum())

# Fill missing Age with mean
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Fill missing Score with median
df['Score'].fillna(df['Score'].median(), inplace=True)

print("\nAfter filling nulls:\n", df.isnull().sum())
print(df)
```

### **Sorting and Filtering**

```python
# Sort by Salary descending
print(df.sort_values('Salary', ascending=False)[['Name','Salary']])

# Filter IT department
print("\nIT Department:\n", df[df['Department'] == 'IT'])

# Value counts
print("\nDepartment Count:\n", df['Department'].value_counts())
```

### **Merging DataFrames**

```python
df1 = pd.DataFrame({'EmpID':[1,2,3,4], 'Name':['Alice','Bob','Charlie','David']})
df2 = pd.DataFrame({'EmpID':[1,2,3,5], 'Department':['HR','IT','Finance','IT']})

inner = pd.merge(df1, df2, on='EmpID', how='inner')
left  = pd.merge(df1, df2, on='EmpID', how='left')

print("Inner Merge:\n", inner)
print("\nLeft Merge:\n", left)
```
