### **Import**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

print("Matplotlib imported!")
```

### **Line Plot**

```python
months  = ['Jan','Feb','Mar','Apr','May','Jun']
sales_a = [1500, 1800, 1700, 2100, 2400, 2200]
sales_b = [1200, 1500, 1600, 1900, 2000, 2300]

plt.figure(figsize=(8,4))
plt.plot(months, sales_a, marker='o', color='blue',  label='Product A')
plt.plot(months, sales_b, marker='s', color='green', label='Product B')
plt.title('Monthly Sales Comparison')
plt.xlabel('Month')
plt.ylabel('Sales (INR)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### **Bar Chart**

```python
departments = ['HR', 'IT', 'Finance', 'Marketing', 'Sales']
avg_salary  = [45000, 62000, 58000, 50000, 47000]

plt.figure(figsize=(8,4))
bars = plt.bar(departments, avg_salary,
               color=['#4C72B0','#DD8452','#55A868','#C44E52','#8172B2'])
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Salary (INR)')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 500,
             f'{int(bar.get_height()):,}',
             ha='center', fontsize=9)
plt.tight_layout()
plt.show()
```

### **Histogram**

```python
np.random.seed(42)
ages = np.random.normal(loc=35, scale=10, size=300).astype(int)
ages = np.clip(ages, 18, 65)

plt.figure(figsize=(8,4))
plt.hist(ages, bins=15, color='steelblue', edgecolor='black', alpha=0.8)
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.axvline(ages.mean(), color='red', linestyle='--', label=f'Mean: {ages.mean():.1f}')
plt.legend()
plt.tight_layout()
plt.show()
```

### **Scatter Plot**

```python
np.random.seed(0)
experience = np.random.randint(1, 20, 50)
salary     = experience * 3000 + np.random.randint(-5000, 5000, 50)

plt.figure(figsize=(8,4))
plt.scatter(experience, salary, color='darkorange', edgecolors='black', alpha=0.7, s=80)
plt.title('Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary (INR)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### **Pie Chart**

```python
labels  = ['HR', 'IT', 'Finance', 'Marketing', 'Sales']
sizes   = [15, 30, 20, 20, 15]
explode = (0, 0.1, 0, 0, 0)

plt.figure(figsize=(7,5))
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'],
        shadow=True, startangle=140)
plt.title('Employee Distribution by Department')
plt.tight_layout()
plt.show()
```

### **Subplots**

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Dashboard - All Plot Types', fontsize=14, fontweight='bold')

# Line
axes[0,0].plot([1,2,3,4,5],[10,25,15,30,20], marker='o', color='blue')
axes[0,0].set_title('Line Plot')
axes[0,0].set_xlabel('X')
axes[0,0].set_ylabel('Y')

# Bar
axes[0,1].bar(['A','B','C','D'],[40,70,55,85],
              color=['#4C72B0','#DD8452','#55A868','#C44E52'])
axes[0,1].set_title('Bar Chart')
axes[0,1].set_xlabel('Category')
axes[0,1].set_ylabel('Value')

# Histogram
data = np.random.randn(500)
axes[1,0].hist(data, bins=20, color='purple', edgecolor='white', alpha=0.8)
axes[1,0].set_title('Histogram')
axes[1,0].set_xlabel('Value')
axes[1,0].set_ylabel('Frequency')

# Scatter
x = np.random.rand(50)
y = x + np.random.rand(50) * 0.3
axes[1,1].scatter(x, y, color='green', edgecolors='black', alpha=0.7)
axes[1,1].set_title('Scatter Plot')
axes[1,1].set_xlabel('X')
axes[1,1].set_ylabel('Y')

plt.tight_layout()
plt.show()
```
