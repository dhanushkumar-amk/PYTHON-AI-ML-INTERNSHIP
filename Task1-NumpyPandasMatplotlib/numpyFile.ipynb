### **Import**

```python
import numpy as np
print("NumPy version:", np.__version__)
```

### **Creating Arrays**

```python
# 1D Array
a = np.array([1, 2, 3, 4, 5])
print("1D Array:", a)

# 2D Array
b = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", b)

# 3D Array
c = np.array([[[1,2],[3,4]], [[5,6],[7,8]]])
print("3D Array:\n", c)

print("Shape of 2D:", b.shape)
print("Dtype:", a.dtype)
print("Dimensions:", c.ndim)
```

### **Built-in Functions**

```python
print("Zeros:\n",       np.zeros((3,3)))
print("Ones:\n",        np.ones((2,4)))
print("Arange:",        np.arange(0, 20, 2))
print("Linspace:",      np.linspace(0, 1, 6))
print("Identity:\n",    np.eye(3))
print("Random Int:\n",  np.random.randint(1, 100, (3,3)))
print("Random Float:\n",np.random.rand(3,3).round(2))
```

### **Array Operations**

```python
x = np.array([10, 20, 30, 40, 50])

print("Sum:",         np.sum(x))
print("Mean:",        np.mean(x))
print("Median:",      np.median(x))
print("Std Dev:",     np.std(x))
print("Variance:",    np.var(x))
print("Min:",         np.min(x), " Max:", np.max(x))
print("Cumsum:",      np.cumsum(x))
print("Squared:",     x ** 2)
print("Square Root:", np.sqrt(x))
print("Log:",         np.log(x).round(2))
```

### **Indexing, Slicing, Reshaping**

```python
arr = np.arange(1, 13).reshape(3, 4)
print("Matrix:\n", arr)
print("Row 0:", arr[0])
print("Col 2:", arr[:, 2])
print("Element [1][2]:", arr[1][2])
print("Slice [0:2, 1:3]:\n", arr[0:2, 1:3])
print("Flattened:", arr.flatten())
```

### **Boolean Masking**

```python
data = np.array([5, 15, 25, 35, 45, 55])
print("Values > 20:",    data[data > 20])
print("Even numbers:",   data[data % 10 == 5])
print("Between 15-45:",  data[(data >= 15) & (data <= 45)])
```

### **Matrix Operations**

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("Addition:\n",       A + B)
print("Multiplication:\n", A * B)
print("Dot Product:\n",    np.dot(A, B))
print("Transpose:\n",      A.T)
print("Determinant:",      round(np.linalg.det(A), 2))
print("Inverse:\n",        np.linalg.inv(A).round(2))
```
